# [The first program](https://nim-lang.org/docs/tut1.html#the-first-program)

They're going right in with an interactive hello world:

```nim
# This is a comment
echo "What's your name? "
var name: string = readLine(stdin)
echo "Hi, ", name, "!"
```

First impressions:
 * Comments are using '#' like Python or Shell. 
    - This seems common for scripting languages?
 * The print command is just `echo`.
    - Very shell like.
 * No brackets around the `echo`.
    - Well... autohotkey does it. Is this actually just a shell?
 * `var` for defining variables, colon typing style.
    - Explict typed declarations are surprising given the loose syntax.
 * `string` is native.
    - No char arrays here!
 * Commas used in string building.
    - Or not... are they seperate parameters?
 * Standard functions exist already without an import.
    - Probably for the best, having import your standard library every time just adds boilerplate.
 * camelCase is used
    - Interesting... I haven't seem many languages explictly use this lately.
 * Brackets *are* used for readLine.
    - What's the rule here???
 * `stdin` is an object
    - How would I access its properties?

So it feels very... easy on the eyes. And the hands.

There's no excessive brackets or magic symbols.

I save the greeting program to 'src/greetings.nim' and ran it with:
```sh
nim compile --run src/greetings.nim
```

Apparently any parameters would be added to the end.

It took a moment to configure the toolchain, then generated and ran the program.

So this is is the debug version which is described as having:
> a large number of runtime checks aiming for your debugging pleasure

Aw yeah. Grovy baby!

Seems good to have a lot of protection by default. 

But you can also add '-d:release' for the release version or '-d:danger' to go without these protections if performance is important. 

They later noted that there is type inference, so reading the name variable could have been done like this:

```nim
var name = readLine(stdin)
```

Also the builtins seen here are from the 'system' module which is implicitly imported.
